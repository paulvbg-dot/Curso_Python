## 2) Variables y tipos

Una **variable** guarda un valor. Python infiere el tipo autom√°ticamente.

Vamos a crear variables de distintos tipos:
- entero (`int`)
- decimal (`float`)
- texto (`str`)
- booleano (`bool`)

```python
edad = 21
promedio = 9.25
nombre = "Paul"
activo = True

print(edad, type(edad))
print(promedio, type(promedio))
print(nombre, type(nombre))
print(activo, type(activo))
```

## 3) Operaciones b√°sicas

- Suma, resta, multiplicaci√≥n, divisi√≥n
- Potencia con `**`
- Divisi√≥n entera con `//`
- Residuo con `%`

```python
a = 17
b = 5

print("a+b =", a+b)
print("a-b =", a-b)
print("a*b =", a*b)
print("a/b =", a/b)
print("a**b =", a**b)
print("a//b =", a//b)
print("a%b =", a%b)
```

## 4) Texto (strings) y concatenaci√≥n

Hay varias formas de construir texto:
- concatenaci√≥n con `+`
- f-strings (recomendado): `f"Hola {nombre}"`

```python
nombre = "Ana"
apellidos = "Bravo"

# Concatenaci√≥n
saludo1 = "Hola " + nombre + " " + apellidos
print(saludo1)

# f-string (recomendado)
saludo2 = f"Hola {nombre} {apellidos}"
print(saludo2)
```
