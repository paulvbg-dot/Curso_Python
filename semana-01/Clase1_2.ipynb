## 2) Conversi√≥n de tipos: `int()` y `float()`

Cuando pides n√∫meros con `input()`, llegan como texto.  
Por eso, para sumar/multiplicar, hacemos conversi√≥n.

```python
edad_txt = input("Dame tu edad: ")
edad = int(edad_txt)          # convertimos a entero
print(f"El pr√≥ximo a√±o tendr√°s {edad + 1} a√±os.")
```

```python
altura_txt = input("Dame tu estatura en metros (ej. 1.72): ")
altura = float(altura_txt)    # convertimos a decimal
print(f"Tu estatura es {altura} m.")
```

## 3) Operadores y precedencia

Recuerda:
- potencia: `**`
- multiplicaci√≥n/divisi√≥n: `*` y `/`
- suma/resta: `+` y `-`

Usa par√©ntesis si quieres forzar un orden.

```python
print("Sin par√©ntesis:", 2 + 3 * 4)
print("Con par√©ntesis:", (2 + 3) * 4)
print("Potencia:", 2 ** 5)
```

## 4) `round()` para redondear

`round(x, 2)` redondea a 2 decimales (√∫til para promedios, dinero, etc.)

```python
precio = 199.9999
print("Precio redondeado:", round(precio, 2))
```

## 5) `len()` para contar caracteres

`len(texto)` da la longitud (n√∫mero de caracteres).

```python
frase = input("Escribe una frase corta: ")
print(f"Tu frase tiene {len(frase)} caracteres.")
```

## 6) Mini-ejemplo completo: conversi√≥n de temperatura

F√≥rmula:  
\[
F = \frac{9}{5}C + 32
\]

```python
c = float(input("Temperatura en ¬∞C: "))
f = (9/5)*c + 32
print(f"{c} ¬∞C equivalen a {round(f, 2)} ¬∞F")
```
