# Formato de salida con f-strings

Esta sección muestra **cómo presentar resultados de forma clara y profesional**
usando **f-strings** en Python.

Las f-strings **no cambian el valor del número**, solo **cómo se muestra**.

## Estructura general

En una f-string, el formato se escribe así:

```
{valor:formato}
```

Esto permite controlar decimales, alineación, porcentaje, etc.

## Regla básica: decimales

```
.nf  →  n decimales
```

Ejemplos:
- `.1f` → 1 decimal  
- `.2f` → 2 decimales  
- `.3f` → 3 decimales

````python
x = 3.14159
print(f"x con 1 decimal: {x:.1f}")
print(f"x con 2 decimales: {x:.2f}")
print(f"x con 3 decimales: {x:.3f}")
````

## Ancho fijo y alineación

Podemos controlar el espacio que ocupa un valor:

- `<`  izquierda  
- `>`  derecha (por defecto)  
- `^`  centrado

````python
print(f"|{42:5}|")
print(f"|{'Ana':<5}|")
print(f"|{'Ana':^5}|")
print(f"|{'Ana':>5}|")
````

## Combinar ancho y decimales

Muy útil para imprimir tablas y reportes:

````python
y = 3.14159
print(f"|{y:8.2f}|")
````

## Porcentajes

Python multiplica por 100 automáticamente:

````python
p = 0.237
print(f"{p:.1%}")
print(f"{p:.2%}")
````

## Notación científica

Para números muy grandes o muy pequeños:

````python
z = 0.000012345
print(f"{z:.2e}")
````

## Separador de miles

````python
n = 123456789
print(f"{n:,}")
````

## Idea clave

> **El formato cambia cómo se ve el número, no el número en sí.**

Estas herramientas ayudan a que los resultados sean más claros y profesionales.
