## 1) √çndices y slicing (rebanadas)

En Python, las cadenas se indexan desde 0.

Ejemplo: `s = "Python"`
- `s[0]` es `'P'`
- `s[-1]` es el √∫ltimo caracter
- `s[1:4]` toma del √≠ndice 1 al 3

```python
s = "Python"
print("s =", s)
print("s[0] =", s[0])
print("s[-1] =", s[-1])
print("s[1:4] =", s[1:4])
print("s[:3] =", s[:3])
print("s[3:] =", s[3:])
```

## 2) M√©todos √∫tiles

- `lower()` / `upper()` cambia a min√∫sculas/may√∫sculas
- `strip()` quita espacios al inicio y al final
- `replace(a, b)` reemplaza partes del texto

```python
texto = "   Hola Mundo   "
print("Original:", repr(texto))
print("strip:", repr(texto.strip()))
print("upper:", texto.strip().upper())
print("lower:", texto.strip().lower())
print("replace:", texto.strip().replace("Mundo", "Python"))
```

## 3) Pertenencia y b√∫squeda

- `"ana" in "bananas"` es `True`
- `find()` regresa el √≠ndice donde aparece (o -1 si no est√°)

```python
t = "bananas"
print("'ana' in t:", "ana" in t)
print("t.find('ana'):", t.find("ana"))
print("t.find('xyz'):", t.find("xyz"))
```

## 4) f-strings para imprimir bonito

Puedes meter variables dentro de strings con `f"..."`.

```python
nombre = "Ana"
edad = 30
print(f"{nombre} tiene {edad} a√±os.")
```

## 5) Mini-ejemplo: limpiar un nombre

Pedir un nombre, quitar espacios, y poner la primera letra en may√∫scula
(*usaremos* `strip()` y `title()`).

```python
nombre = input("Escribe tu nombre (con espacios si quieres): ")
limpio = nombre.strip().title()
print(f"Nombre limpio: {limpio}")
```
