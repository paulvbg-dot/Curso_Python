## 1) Comparaciones

Las comparaciones devuelven `True` o `False`.

```python
print(3 < 5)
print(10 == 10)
print(7 != 2)
print(4 >= 4)
```

## 2) `if` / `else`

Ejemplo: decir si una persona es mayor de edad.

```python
edad = int(input("Edad: "))
if edad >= 18:
    print("Eres mayor de edad.")
else:
    print("Eres menor de edad.")
```

## 3) `elif` (varias opciones)

Ejemplo: calificaci√≥n en letras (un ejemplo sencillo).

```python
cal = float(input("Calificaci√≥n (0 a 10): "))

if cal >= 9:
    print("Excelente")
elif cal >= 7:
    print("Bien")
elif cal >= 6:
    print("Suficiente")
else:
    print("Insuficiente")
```

## 4) Operadores l√≥gicos: `and`, `or`, `not`

- `and` exige que ambas condiciones sean verdaderas
- `or` basta con que una sea verdadera
- `not` invierte la condici√≥n

```python
x = int(input("x: "))
y = int(input("y: "))

if x > 0 and y > 0:
    print("Ambos son positivos.")
elif x > 0 or y > 0:
    print("Al menos uno es positivo.")
else:
    print("Ninguno es positivo.")
```

## 5) Validaci√≥n simple

Si esperamos un valor en cierto rango, conviene revisar.

```python
cal = float(input("Calificaci√≥n (0 a 10): "))
if 0 <= cal <= 10:
    print("Calificaci√≥n v√°lida.")
else:
    print("Calificaci√≥n inv√°lida. Debe estar entre 0 y 10.")
```
