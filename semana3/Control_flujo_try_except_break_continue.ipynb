
# Control de flujo y manejo de errores en Python

En este notebook veremos palabras clave importantes del mismo estilo:

## üî¥ Manejo de errores
- `try`
- `except`
- `else`
- `finally`
- `raise`

## üîµ Control de ciclos
- `break`
- `continue`
- `pass`

Cada secci√≥n tiene ejemplos peque√±os y claros para practicar.



## 1Ô∏è‚É£ `try` y `except`

Sirven para evitar que el programa se rompa cuando ocurre un error.


```

try:
    x = int("hola")  # esto genera un error
except ValueError:
    print("‚ùå No se pudo convertir a entero")
```


## 2Ô∏è‚É£ `else`

El bloque `else` se ejecuta solo si **NO hubo error** dentro del `try`.


```

try:
    x = int("5")
except ValueError:
    print("Error")
else:
    print("‚úÖ Conversi√≥n correcta:", x)
```


## 3Ô∏è‚É£ `finally`

`finally` se ejecuta **siempre**, haya error o no.


```

try:
    print("Intentando algo...")
finally:
    print("üßπ Esto siempre se ejecuta")
```


## 4Ô∏è‚É£ `raise`

Sirve para lanzar un error manualmente.


```

x = -3

if x < 0:
    raise ValueError("‚ùå El n√∫mero no puede ser negativo")
```


## 5Ô∏è‚É£ `break`

Rompe completamente un ciclo `while` o `for`.


```

i = 0
while True:
    print(i)
    if i == 3:
        break
    i += 1
```


## 6Ô∏è‚É£ `continue`

Salta a la siguiente iteraci√≥n del ciclo.


```

for i in range(5):
    if i == 2:
        continue
    print(i)
```


## 7Ô∏è‚É£ `pass`

No hace nada. Se usa como marcador cuando a√∫n no escribes el c√≥digo.


```

for i in range(3):
    if i == 1:
        pass
    print(i)
```


## 8Ô∏è‚É£ Ejemplo combinado

Aqu√≠ juntamos varias palabras clave:
- `while`
- `try`
- `except`
- `continue`
- `break`


```

while True:
    s = input("Escribe un n√∫mero (o 'ok' para salir): ").strip().lower()
    
    if s == "ok":
        print("üëã Saliendo...")
        break
    
    try:
        n = int(s)
    except ValueError:
        print("‚ùå No es n√∫mero, intenta otra vez")
        continue
    
    print("‚úÖ N√∫mero v√°lido:", n)
```
